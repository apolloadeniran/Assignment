###Problem 1 Implementation using exponentiation arithmetic *operators*

```python
THICKNESS = 0.00008

folded_thickness = THICKNESS * 2**43

print("Thickness: {} meters".format(folded_thickness))
```

###Problem 2 Unit Conversion

```python
print("Thickness: {: .3f} kilometers".format(folded_thickness / 1000))
```

###Problem 3 Create using a for statement

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(1, 44):
 folded_thickness *= 2
print("Thickness: {} meters".format(folded_thickness))
```

###Problem 4 Comparison of calculation time

```python
import time
start = time.time()
THICKNESS = 0.00008

folded_thickness = THICKNESS * 2**43

print("Thickness: {} meters".format(folded_thickness))
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
%%timeit
THICKNESS = 0.00008

folded_thickness = THICKNESS * 2**43

print("Thickness: {} meters".format(folded_thickness))
```

###Problem 5 Saving to *a* list

```python
folder_thickness_list = []
folder_thickness = THICKNESS
for i in range (1, 44):
 folder_thickness *= 2
 folder_thickness_list.append(folder_thickness)
```

###Problem 6 Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folder_thickness_list)  # Enter the variable name of the list in "List name"
plt.show()
```

###Problem 7 Customizing graphs

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folder_thickness_list, color='red')
plt.show()
```
