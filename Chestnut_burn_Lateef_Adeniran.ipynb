###Problem. The day when chestnut buns cover the solar system

```python
import matplotlib.pyplot as plt
%matplotlib inline
import math

CHESTNUT_DIAMETER_CM = 2.54
CHESTNUT_RADIUS_CM = CHESTNUT_DIAMETER_CM / 2
CHESTNUT_VOLUME = 4/3 * math.pi *(CHESTNUT_RADIUS_CM ** 3)
CHESTNUT_VOLUME_M3 = CHESTNUT_VOLUME_CM3 = 1e-6

SOLAR_SYSTEM_RADIUS_METERS = 4.545e9
SOLAR_SYSTEM_VOLUME_M3 = (4/3) * math.pi * (SOLAR_SYSTEM_RADIUS_METERS ** 3)
doubling_time = 0

volume = []
minute = []
def calculate_doubling_time(initial_volume, target_volume):
    doubling_time = 0
    while initial_volume < target_volume:
        initial_volume*= 2
        volume.append(initial_volume)
        doubling_time += 5
        minute.append(doubling_time)
    return doubling_time
time_to_fill_solar_system = calculate_doubling_time(CHESTNUT_VOLUME_M3, SOLAR_SYSTEM_VOLUME_M3)
print(f"it will take (time_to_fill_solar_system []th) minutes to cover the solar system with chestnut burn.")

plt.xlabel("Time(minutes)")
plt.ylabel("volume(cubic meters)")
plt.plot(minute, volume)
plt.title("chestnut burn volume increase over time")
plt.grid(True)
plt.show()
```
